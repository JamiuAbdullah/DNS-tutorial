## Exercise 5 — Loops
prices = [500,1200,350,2000,150]
1. Print each price
2. Print prices > 1000
3. Count prices <= 500

```python
prices = [500,1200,350,2000,150]

      # Print each price
for price in prices:
    print(price)

     # Print prices > 1000
for price in prices:
    if price > 1000:
        print("price greater than 1000 =", price)

      # Count prices <= 500
count = 0
for price in prices:
    if price <= 500:
        count += 1

print("count of price less than or equal to 500 =",count)
```

## Final Lab — Inventory Management System
inventory = {'rice':10,'beans':5,'milk':8}
Build a menu program to manage inventory.

```python
inventory = {'rice': 10, 'beans': 5, 'milk': 8}

         # Function to display current inventory
def show_inventory():
    print("\nCurrent Inventory:")
    if inventory:
        for item, quantity in inventory.items():
            print(f"{item}: {quantity}")
    else:
        print("Inventory is empty.")
    print()

          # Function to add a new item or increase quantity of existing item
def add_item():
    item = input("Enter item name to add: ").lower()
    quantity = int(input("Enter quantity: "))
    if item in inventory:
        inventory[item] += quantity
        print(f"{quantity} added to {item}. New quantity: {inventory[item]}")
    else:
        inventory[item] = quantity
        print(f"{item} added to inventory with quantity {quantity}.")

          # Function to remove an item from inventory
def remove_item():
    item = input("Enter item name to remove: ").lower()
    if item in inventory:
        del inventory[item]
        print(f"{item} has been removed from inventory.")
    else:
        print(f"{item} not found in inventory.")

            # Function to update quantity of an existing item
def update_item():
    item = input("Enter item name to update: ").lower()
    if item in inventory:
        quantity = int(input("Enter new quantity: "))
        inventory[item] = quantity
        print(f"{item} quantity updated to {quantity}.")
    else:
        print(f"{item} not found in inventory.")

def main_menu():
    while True:
        
        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            show_inventory()
        elif choice == '2':
            add_item()
        elif choice == '3':
            remove_item()
        elif choice == '4':
            update_item()
        elif choice == '5':
            print("Exiting program. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number from 1 to 5.")

main_menu()
```
